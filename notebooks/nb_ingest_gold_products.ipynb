```python
from pyspark.sql import functions as F
```

```python


df_stage_product = (spark.table("stage.stg_new_products")
    .select(
        "product_id",
        "product_name",
        "category",
        "unit_price"
    )
)

df_gold_product = (df_stage_product
    .filter(F.col("product_id").isNotNull())
    .dropDuplicates(["product_id"])
    .withColumn("unit_price", F.col("unit_price").cast("double"))
    .withColumn("created_date", F.current_timestamp())
    .withColumn("dim_product_key", F.monotonically_increasing_id())
)


(df_gold_product.write
    .mode("append")
    .format("delta")
    .saveAsTable("gold.dim_product")
)

print("Load data into gold.dim_product completed successfully")
```
