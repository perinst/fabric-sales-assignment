```python
from pyspark.sql import functions as F
```

Load CSVs from Lakehouse Files

```python
df_customers = spark.read.option("header", True).option("inferSchema", True) \
    .csv("Files/raw_uploads/customers.csv")

df_products = spark.read.option("header", True).option("inferSchema", True) \
    .csv("Files/raw_uploads/products.csv")

df_transactions = spark.read.option("header", True).option("inferSchema", True) \
    .csv("Files/raw_uploads/transactions.csv")
```

#  Add ingestion metadata 

```python
from datetime import datetime
ingested_at = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

df_customers    = df_customers.withColumn("ingested_at",    F.lit(ingested_at))
df_products     = df_products.withColumn("ingested_at",     F.lit(ingested_at))
df_transactions = df_transactions.withColumn("ingested_at", F.lit(ingested_at))
```

#  Write as Delta tables (Raw zone) 

```python
df_customers.write.mode("overwrite").format("delta") \
    .saveAsTable("raw_customers")

df_products.write.mode("overwrite").format("delta") \
    .saveAsTable("raw_products")

df_transactions.write.mode("overwrite").format("delta") \
    .saveAsTable("raw_transactions")

print(f"  raw_customers:    {df_customers.count()} rows")
print(f"  raw_products:     {df_products.count()} rows")
print(f"  raw_transactions: {df_transactions.count()} rows")
```
