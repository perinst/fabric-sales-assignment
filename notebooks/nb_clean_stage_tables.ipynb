```python
from pyspark.sql import functions as F

stage_schema = "stage"

print(f"Scanning schema : {stage_schema}")


df_tables = spark.sql(f"SHOW TABLES IN {stage_schema}")

stage_tables = [
    row.tableName
    for row in df_tables.collect()
]

print(f"Found {len(stage_tables)} stage tables")


for table_name in stage_tables:

    full_table_name = f"{stage_schema}.{table_name}"

    try:

        print(f"Cleaning table : {full_table_name}")

        spark.sql(f"TRUNCATE TABLE {full_table_name}")

        print(f"SUCCESS : {full_table_name}")

    except Exception as e:

        print(f"FAILED : {full_table_name}")
        print(str(e))


print("\nValidation after cleanup")

for table_name in stage_tables:

    full_table_name = f"{stage_schema}.{table_name}"

    row_count = spark.table(full_table_name).count()

    print(f"{full_table_name} = {row_count} rows")

print("\nStage cleanup completed")
```
