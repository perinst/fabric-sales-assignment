```python
spark.sql("CREATE SCHEMA IF NOT EXISTS audit")

spark.sql("""
    CREATE TABLE IF NOT EXISTS audit.pipeline_log (
        run_id          STRING,
        file_name       STRING,
        entity          STRING,
        load_type       STRING,
        rows_read       LONG,
        rows_inserted   LONG,
        rows_updated    LONG,
        status          STRING,
        error_message   STRING,
        start_time      TIMESTAMP,
        end_time        TIMESTAMP
    )
    USING DELTA
""")

print("audit.pipeline_log created")
```
